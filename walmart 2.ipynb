```python
import pandas as pd
import numpy as np
import os,sys
```

```python
sell_prices = pd.read_csv('sell_prices.csv')
validation = pd.read_csv('sales_train_validation.csv')
calendar = pd.read_csv('calendar.csv')
submit = pd.read_csv('sample_submission.csv')
validation
```

```python
from sklearn.neural_network import MLPRegressor
clf = MLPRegressor(activation='relu', alpha=1e-05, batch_size='auto', beta_1=0.9,
       beta_2=0.999, early_stopping=True, epsilon=1e-08,
       hidden_layer_sizes=(5, 2), learning_rate='constant',
       learning_rate_init=0.001, max_iter=1000, momentum=0.9,
       nesterovs_momentum=True, power_t=0.5, random_state=1, shuffle=True,
       solver='lbfgs', tol=0.0001, validation_fraction=0.1, verbose=False,
       warm_start=False)
X_train=validation[validation.columns[(1891-7):1891]]
y_train=validation[validation.columns[1891:1919]]
X_test=validation[validation.columns[(1919-7):1919]]
clf.fit(X_train,y_train)

#X_train
#y_train
list_of_arrays = clf.predict(X_test)

output_test = pd.DataFrame(list_of_arrays).round(0)
output_test
```

```python
#createValidation
validationSubmit = pd.DataFrame()
validationSubmit['id'] = validation['id']
for i in range(1,29):
    validationSubmit['F'+str(i)]=y_train['d_'+str(1886+(i-1))]
validationSubmit

#create evaluation
evaluationSubmit = pd.DataFrame()
evaluationSubmit['id'] = validation['id']
evaluationSubmit = evaluationSubmit.replace(regex=['validation'],value=['evaluation'])
for i in range(1,29):
    evaluationSubmit['F'+str(i)]=output_test[(i-1)]
evaluationSubmit

#create and append Naive algorithm submission
NN_with28outputs = validationSubmit
NN_with28outputs = NN_with28outputs.append(evaluationSubmit, ignore_index=True)
NN_with28outputs

NN_with28outputs.to_csv('NN_with28outputs.csv',index=False)
```
